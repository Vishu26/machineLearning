```python
# Neural Network to learn XOR function

import numpy as np

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])
```

```python
batch_size = 4
n_inputs = 2
n_hidden = 2
n_outputs = 1
n_epochs = 50000
```

```python
X_b = np.c_[np.ones((batch_size, 1)), X] # Adding bias
W_1 = np.random.randn(n_inputs+1, n_hidden) # First Layer Weights
W_2 = np.random.randn(n_hidden, n_outputs) # Final Layer Weights
```

```python
def sigmoid(X, deriv=False):
    if not deriv:
        return 1/(1+np.exp(-X))
    return X*(1-X)
```

```python
# training

for epoch in range(n_epochs):
    
    #### Forward Propagation ####
    layer1 = sigmoid(X_b.dot(W_1))
    layer2 = sigmoid(layer1.dot(W_2))
    
    error_layer_2 = layer2 - y
    if epoch%100==0:
        print('Error :', np.mean(error_layer_2**2))
        
    #### Back Propagation ####
    layer2_delta = error_layer_2*sigmoid(layer2, deriv=True)
    error_layer_1 = layer2_delta.dot(W_2.T)
    layer1_delta = error_layer_1*sigmoid(layer1, deriv=True)
    
    #### Update ####
    
    W_1 = W_1 - X_b.T.dot(layer1_delta)*0.1
    W_2 = W_2 - layer1.T.dot(layer2_delta)*0.1
```

```python
layer1 = sigmoid(np.array([[1, 0, 1]]).dot(W_1))  # prediction
layer2 = sigmoid(layer1.dot(W_2))
print(layer2)                  # Answer close to 1
```

```python
layer1 = sigmoid(np.array([[1, 0, 0]]).dot(W_1)) # prediction
layer2 = sigmoid(layer1.dot(W_2))
print(layer2)                  # Answer close to zero
```
